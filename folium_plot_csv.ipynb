```python
fname = 'foodshed.csv' #this file is created by running gpx_to_csv.ipynb
f = pd.read_table(fname, 
                          header = 0,
                          sep    = ',',
                          names  = ['name','lat','lon','source','p1','p2','p3']
                         );
```

```python
tableau20 = [(31, 119, 180), (174, 199, 232), (255, 127, 14), (255, 187, 120),    
             (44, 160, 44), (152, 223, 138), (214, 39, 40), (255, 152, 150),    
             (148, 103, 189), (197, 176, 213), (140, 86, 75), (196, 156, 148),    
             (227, 119, 194), (247, 182, 210), (127, 127, 127), (199, 199, 199),    
             (188, 189, 34), (219, 219, 141), (23, 190, 207), (158, 218, 229)]
colors=[]
for tab in tableau20:
    colors.append('#%02x%02x%02x' % tab)
```

```python
pname = 'addend.csv' #this file is created by running gpx_to_csv.ipynb
p = pd.read_table(pname, 
                          header = 0,
                          sep    = ',',
                          names  = ['address','lat','lon','source']
                         );
```

```python
icon_image = 'http://leafletjs.com/examples/custom-icons/leaf-red.png'
shadow_image = 'http://leafletjs.com/examples/custom-icons/leaf-shadow.png'

icon = CustomIcon(
    icon_image,
    icon_size=(38, 95),
    icon_anchor=(22, 94),
    shadow_image=shadow_image,
    shadow_size=(50, 64),
    shadow_anchor=(4, 62),
    popup_anchor=(-3, -76)
)
```

```python
ZOOM_LEVEL = 16
FOLIUM_LINE_OPTIONS = dict(opacity=1.0, weight=4)

#get a central GPS location over which to generate a map
lat_avg = f.lat.mean()
lon_avg = f.lon.mean()

#center=[lat_avg,lon_avg]
#center=[42.463598, -70.946082]
center=[42.467249, -70.941656]
map_osm = folium.Map(location=center,zoom_start=ZOOM_LEVEL)

folium.TileLayer('cartodbpositron').add_to(map_osm)


for i in range(0,len(f.lat)):
    lat = f.lat[i]
    lon = f.lon[i]
    source=f.source[i]
    color='blue'
    if source.strip() == 'Cafe':
        color=colors[0]
    if source.strip() == 'Corner Store/ Bodega':
        #color=colors[1]
        color='blue'
    if source.strip() == 'Drug/Discount Store':
        color=colors[2]
    if source.strip() == 'Gas Station - Snack Retail':
        color=colors[3]
    if source.strip() == 'Full Service':
        color=colors[4]
    if source.strip() == 'Liquor Store - Snack Retail':
        color=colors[5]
    if source.strip() == 'Major Franchise Fast Food':
        color=colors[6]
    if source.strip() == 'Supermarket':
        color=colors[7]
    if source.strip() == 'Take Out':
        color=colors[8]
    pt = [lat,lon]
    #folium.Marker(pt, popup='Mt. Hood Meadows',icon = folium.Icon(color='green')).add_to(map_osm)
    #folium.Marker(pt, popup='Mt. Hood Meadows',icon = icon).add_to(map_osm)
    folium.CircleMarker(location=pt, radius=6, color=color,fill_color=color,popup=source).add_to(map_osm)

for i in range(0,len(p.lat)):
    lat = p.lat[i]
    lon = p.lon[i]
    source=p.source[i]
    pt = [lat,lon]
    if source.strip() == 'TFP Office & Farm':
        color='green'
        folium.RegularPolygonMarker(pt, popup=source, fill_color=color,number_of_sides=3,radius=14).add_to(map_osm)
    if source.strip() == 'TFP Ingalls School Farm':
        color='green'
        folium.RegularPolygonMarker(pt, popup=source, fill_color=color,number_of_sides=3,radius=14).add_to(map_osm)
    if source.strip() == 'Central Square Farmers Market':
        color='green'
        folium.RegularPolygonMarker(pt, popup=source, fill_color=color,number_of_sides=3,radius=14).add_to(map_osm)
    
    if source.strip() == 'Hunger Relief Organization':
        color='blue'
        folium.RegularPolygonMarker(pt, popup=source, fill_color=color,number_of_sides=4,radius=10).add_to(map_osm)
 
    if source.strip() == 'School':
        color='orange'
        folium.RegularPolygonMarker(pt, popup=source, fill_color=color,number_of_sides=4,radius=10).add_to(map_osm)
    if source.strip() == 'School Garden':
        color='blue'
        folium.RegularPolygonMarker(pt, popup=source, fill_color=color,number_of_sides=4,radius=10).add_to(map_osm)
    if source.strip() == 'School Center':
        color='blue'
        folium.RegularPolygonMarker(pt, popup=source, fill_color=color,number_of_sides=4,radius=10).add_to(map_osm)

    if source.strip() == 'WIC':
        color='blue'
        folium.RegularPolygonMarker(pt, popup=source, fill_color=color,number_of_sides=4,radius=10).add_to(map_osm)
    
    if source.strip() == 'Community Garden':
        color='purple'
        folium.RegularPolygonMarker(pt, popup=source, fill_color=color,number_of_sides=4,radius=10).add_to(map_osm)
map_osm
```
